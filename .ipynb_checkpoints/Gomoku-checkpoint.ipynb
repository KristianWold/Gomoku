```python
class Game:
    
    def __init__(self, agents):
        
        self.agents = agents
        
        #self.fig = plt.figure()
        #self.ax = self.fig.add_subplot(111)
        
        #self.fig.show()
        #self.fig.canvas.draw()
   
    def count_in_a_row(self, i, move, x, y):
        
        in_a_row = 1
        
        counting = True
        j, k = move
        
        while(counting and 0 < j < 14 and 0 < k < 14):
            if self.state[i, j+x, k+y] == 1:
                in_a_row += 1
                j += x
                k += y
            else:
                counting = False
        
        counting = True
        j, k = move
        
        while(counting and 0 < j < 14 and 0 < k < 14):
            if self.state[i, j-x, k-y] == 1:
                in_a_row += 1
                j -= x
                k -= y
            else:
                counting = False
            
        return in_a_row

    def check_win(self, i, move):
        j, k = move
        win = False
        
        if self.count_in_a_row(i, move, 1, 0) >= 5:
            win = True
            
        if self.count_in_a_row(i, move, 0, 1) >= 5:
            win = True
        
        if self.count_in_a_row(i, move, 1, 1) >= 5:
            win = True
            
        if self.count_in_a_row(i, move, 1, -1) >= 5:
            win = True
        
        return win
            
        

    def next_move(self):
        i = self.turn % 2
        
        if i == 0:
            move = self.agents[i](self.state)
        else:
            move = self.agents[i](self.state[::-1])

        self.state[i][move] = 1
        
        self.X[i].append(np.copy(self.state))
        self.y[i].append(15*move[0] + move[1])
        
        self.turn += 1
        
        return self.check_win(i, move)
    
    def play(self):
        self.state = np.zeros((2,15,15), dtype = int)
        self.turn = 0
        self.X = [[np.copy(self.state)],[]]
        self.y = [[],[]]
        
        
        while(not self.next_move()):
            pass
        
        
    def plot(self):
        player1 = np.where(self.state[0] == 1)
        player2 = np.where(self.state[1] == 1)

        self.ax.clear()   
        self.ax.plot(player1[0]+0.5, player1[1]+0.5, "ro")
        self.ax.plot(player2[0]+0.5, player2[1]+0.5, "bo")
        plt.xlim((0, 15))
        plt.ylim((0, 15))
        self.fig.canvas.draw()
             

def user_input(states):
    x = int(input())
    y = int(input())
    return (x, y)
```
